# Tokenization

```
! pip install transformers
```

```
import transformers
```

```
# tokenizer = BertTokenizer.from_pretrained('bert-large-uncased', do_lower_case=True)
```

```
tokenizer = transformers.AutoTokenizer.from_pretrained('cardiffnlp/twitter-roberta-base-sentiment-latest')
```

#### finding max_lenght

```
tokenized_train_temp = tokenizer(train.append(test)['text'].tolist(), max_length=10000, 
                         truncation=True, padding=True, 
                         add_special_tokens=True, return_tensors='np')
```

```
tokenized_train_temp
```

```
tokenized_train_temp["input_ids"]
```

```
print(len(tokenized_train_temp["input_ids"][0]))
print(len(tokenized_train_temp["input_ids"][125]))
print(len(tokenized_train_temp["input_ids"][-10]))
```

```
text_vector_len=len(tokenized_train_temp["input_ids"][0])
```

## Tokenization train

```
tokenized_train = tokenizer(train['text'].tolist(), max_length=text_vector_len, 
                         truncation=True, padding='max_length', 
                         add_special_tokens=True, return_tensors='np')
```

```
tokenized_train
```

```
train_input_ids=tokenized_train["input_ids"]
train_attention_mask=tokenized_train["attention_mask"]
```

```
train_input_ids[3]
```

```
train_attention_mask
```

```
y_train = train['target'].values
```

## Tokenization test

```
tokenized_test = tokenizer(test['text'].tolist(), max_length=text_vector_len, 
                         truncation=True, padding='max_length', 
                         add_special_tokens=True, return_tensors='np')
```

```
test_input_ids=tokenized_test["input_ids"]
test_attention_mask=tokenized_test["attention_mask"]
```

```
tokenized_test["input_ids"][-3]
```

# Creating train and validation dataset

```
!pip install datasets
```

```
from datasets import Dataset
```

```
dic = { "input_ids" : train_input_ids, "attention_mask" : train_attention_mask, "label": y_train}
```

```
data_set=Dataset.from_dict(dic)
```

```
data_set=data_set.train_test_split(test_size=0.2)
```

```
data_set
```

```
test_data_dict={"input_ids" : test_input_ids, "attention_mask" : test_attention_mask}
```

```
test_dataset=Dataset.from_dict(test_data_dict)
```

# Building and training the model

```
# from transformers import BertForSequenceClassification
from transformers import DataCollatorWithPadding
from transformers import AutoModelForSequenceClassification,TrainingArguments,Trainer
```

```
training_args = TrainingArguments("test-trainer")
```

```
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```
model = AutoModelForSequenceClassification.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment-latest", num_labels=2,ignore_mismatched_sizes=True)
```

```
trainer = Trainer(
    model,
    training_args,
    train_dataset=data_set["train"],
    eval_dataset=data_set["test"],
    data_collator=data_collator,
    tokenizer=tokenizer,
)
```

```
trainer.train()
```

# Prediction and evaluation

```
import numpy as np
```

```
predictions=trainer.predict(test_dataset)
```

```
preds = np.argmax(predictions.predictions, axis=-1)
```

```
preds
```

```
submission["target"]=preds
```

```
submission.to_csv("/content/drive/MyDrive/Colab Notebooks/dataset/dataset lab 5/submission.csv",index=False)
```

#Clustering

```
from sklearn.cluster import KMeans, AgglomerativeClustering, AffinityPropagation, SpectralClustering, DBSCAN
from sklearn import metrics
```

```
combined_dataset=np.concatenate((train_input_ids,test_input_ids))
```

```
combined_dataset.shape
```

```
km = KMeans(n_clusters=10)
km.fit(combined_dataset)

clusters_km = km.labels_.tolist()
```

```
ac=AgglomerativeClustering()
ac.fit(combined_dataset)

clusters_ac=ac.labels_.tolist()
```

```
combined_df=train.append(test)
```
