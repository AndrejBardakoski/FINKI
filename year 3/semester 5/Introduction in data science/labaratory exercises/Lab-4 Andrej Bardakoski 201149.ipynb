# traint test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data[data.columns[:-1]],  data[' enjoy_watching '], test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

# модели

```python
# Y ќе прима вредности од 0-4 наместо 1-5
Y_train = Y_train-1
```

```python
Y_test=Y_test-1
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
len(X_train[0])
```

```python
model1 = Sequential()

model1.add(Dense(32, input_dim=20, kernel_initializer='uniform', activation='relu'))

model1.add(Dense(128, kernel_initializer='uniform', activation='relu'))
model1.add(Dense(256, kernel_initializer='uniform', activation='relu'))
model1.add(Dense(64, kernel_initializer='uniform', activation='relu'))

model1.add(Dense(5, activation='softmax'))
```

```python
model1.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy']) 
```

```python
# model1.fit(X_train, Y_train, epochs=20, batch_size=1) 
```

```python
# model1.fit(X_train, Y_train, epochs=250, batch_size=12) 
```

```python
# model1.fit(X_train, Y_train, epochs=150, batch_size=8) 
```

```python
model1.fit(X_train, Y_train, epochs=110, batch_size=6) 
```

```python
y_pred1 = model1.predict(X_test) 
```

```python
model2 = Sequential()

model2.add(Dense(32, input_dim=20, kernel_initializer='uniform', activation='relu'))
model2.add(Dropout(0.2, input_shape=(32,)))
model2.add(Dense(64, kernel_initializer='uniform', activation='relu'))
model2.add(Dropout(0.1, input_shape=(64,)))
model2.add(Dense(128, kernel_initializer='uniform', activation='relu'))
model2.add(Dropout(0.1, input_shape=(128,)))
model2.add(Dense(64, kernel_initializer='uniform', activation='relu'))

model2.add(Dense(5, activation='softmax'))
```

```python
model2.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy']) 
```

```python
# model2.fit(X_train, Y_train, epochs=60, batch_size=5) 
```

```python
# model2.fit(X_train, Y_train, epochs=100, batch_size=4) 
```

```python
model2.fit(X_train, Y_train, epochs=200, batch_size=10)
```

```python
y_pred2 = model2.predict(X_test) 
```

```python
model3 = Sequential()

model3.add(Dense(64, input_dim=20, kernel_initializer='uniform', activation='relu'))
model3.add(Dropout(0.3, input_shape=(64,)))

model3.add(Dense(256, kernel_initializer='uniform', activation='relu'))
model3.add(Dropout(0.3, input_shape=(256,)))
model3.add(Dense(256, kernel_initializer='uniform', activation='relu'))
model3.add(Dropout(0.25, input_shape=(256,)))
model3.add(Dense(256, kernel_initializer='uniform', activation='relu'))
model3.add(Dense(512, kernel_initializer='uniform', activation='relu'))
model3.add(Dense(64, kernel_initializer='uniform', activation='relu'))
model3.add(Dropout(0.25, input_shape=(64,)))
model3.add(Dense(32, kernel_initializer='uniform', activation='relu'))
model3.add(Dropout(0.2, input_shape=(32,)))

model3.add(Dense(5, activation='softmax'))
```

```python
model3.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy']) 
```

```python
# model3.fit(X_train, Y_train, epochs=50, batch_size=4) 
```

```python
# model3.fit(X_train, Y_train, epochs=200, batch_size=12) 
```

```python
model3.fit(X_train, Y_train, epochs=130, batch_size=7) 
```

```python
y_pred3 = model3.predict(X_test) 
```

# Evaluation

```python
from sklearn import metrics
```

```python
from sklearn.metrics import classification_report, confusion_matrix, f1_score
```

```python
y_pred1=np.argmax(y_pred1, axis=1)
```

```python
print(classification_report(Y_test, y_pred1))
```

```python
y_pred2=np.argmax(y_pred2, axis=1)
```

```python
print(classification_report(Y_test, y_pred2))
```

```python
y_pred3=np.argmax(y_pred3, axis=1)
```

```python
print(classification_report(Y_test, y_pred3))
```

```python
print("model 1: ",metrics.accuracy_score(Y_test, y_pred1))
print("model 2: ",metrics.accuracy_score(Y_test, y_pred2))
print("model 3: ",metrics.accuracy_score(Y_test, y_pred3))
```
