Таргет колона е rating.

Се однесува на рејтингот (оценката) кој го дал одреден корисник за одреден филм ние ќе пробаме да го предвидиме тој рејтинг со користење на податоците од останатите колони кои се однесуваат на персоналноста на корисникот.

Ова е регресионен проблем затоа што таргет колоната е рејтинг (оценка) кој има нумеричка вредност од 0 до 5, но исто така може да се смета и како категориски проблем затоа што оценката прима само одредени вредности и тоа (0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5)

ќе биде тешко скоро невозможно добро да се предвиди оценката поради тоа што имаме податоци само за корисникот, а за филмот единствена информација е movie_id која што не ни кажува ништо за содржината или квалитетот на филмот (ништо со што би можеле да ја предвидиме оценката добиена од соодветниот корисник)

```python
data_wout_pred_rtg=data.drop(data.columns[range(5,34)],axis = 1)
```

# traint test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data_wout_pred_rtg[data_wout_pred_rtg.columns[:-1]],  data_wout_pred_rtg['rating'], test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

# модели

## линеарна регресија

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit(X_train, Y_train)
```

```python
model.score(X_train, Y_train)
```

```python
Y_pred_lin = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(Y_test, Y_pred_lin)
```

score = 0.007176591398748711 (mnogu los rezultat)

##KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

gi skalirame Y mnozestvata so 2 za da dobieme celi broevi koi sto moze da se iskoristat za klasifikacija

```python
Y_train_x2=Y_train*2
```

```python
Y_test_x2=Y_test*2
```

```python
Y_train_x2
```

```python
classifier = KNeighborsClassifier(n_neighbors=100)
classifier.fit(X_train, Y_train_x2)
```

```python
Y_pred_knn = classifier.predict(X_test) 
```

```python
from sklearn import metrics
```

```python
from sklearn.metrics import classification_report, confusion_matrix, f1_score
```

```python
print(confusion_matrix(Y_test_x2, Y_pred_knn))
```

```python
print(classification_report(Y_test_x2, Y_pred_knn))
```

```python
print("Accuracy:",metrics.accuracy_score(Y_test_x2, Y_pred_knn))
```

добиваме резултат од 0.25 што е лош но имаќи во предвид што предвидуваме и начинот на кој што го правиме тоа и овој резултат не е лош

## Decision Tree classification

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier(max_depth=30)
classifier = classifier.fit(X_train,Y_train_x2)

y_pred_dTree_cf = classifier.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(Y_test_x2, y_pred_dTree_cf))
```

```python
print(confusion_matrix(Y_test_x2, y_pred_dTree_cf))
print(classification_report(Y_test_x2, y_pred_dTree_cf))
```
